```python
# import the pipeline from transformers
from transformers import pipeline
```

### 1) sentiment analysis example

```python
# take a sample text of negative review and guess what will be the output
sample_text = "Hey! I am watching this movie and you recommended it to watch but my guess is that this movie is not worth watching"
classifier = pipeline("sentiment-analysis") # performing sentiment analysis
result = classifier(sample_text)
result
```

### result: the model has printed a negative review with a score of 99% which means that the model is performing very well

### Batch sentiment analysis means text in a list separated by a comma

```python
classifier = pipeline(task = "sentiment-analysis", model = "Samlowe/roberta-base-go_emotions") # here model is also provided
sample_text_lst = [
    "last time when i met you, i felt a deep attraction and love", \
    "But now when i see you i do not feel anything anymore", \
    "The reason is that i wanted you and wanted to marry with you",\
    "but what you did is not bearable at all", \
    "thanks for being in my life for a short period of time"

]
result = classifier(sample_text_lst)
result
```
